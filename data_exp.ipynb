```python
data = pd.read_csv('nhl_202324.csv')
data.head(10)
```

```python
total = len(data)
num_home_wins = data['HW'].values.sum() + data['HS'].values.sum()
num_away_wins = data['AW'].values.sum() + data['AS'].values.sum()
print("Home team won {} out of {} times. ({:%})".format(num_home_wins, total, num_home_wins / total))
print("Away team won {} out of {} times. ({:%})".format(num_away_wins, total, num_away_wins / total))
```

```python
team_win_rates = {}
team_wins_count = {}
team_reg_loses_count = {}
team_so_loses_count = {}
for team in team_abbr:
    if team == 'UTA':
        break
    games = data.loc[(data['AT'] == team) | (data['HT'] == team)]
    wins = data.loc[data['WON'] == team]
    reg_loses = data.loc[data['LOST'] == team]
    so_loses = data.loc[data['SOLOST'] == team]
    team_win_rates[team] = len(wins) / len(games)
    team_wins_count[team] = len(wins)
    team_reg_loses_count[team] = len(reg_loses)
    team_so_loses_count[team] = len(so_loses)
team_records = pd.DataFrame([team_win_rates,team_wins_count,team_reg_loses_count,team_so_loses_count]).T
team_records.columns = ['WIN RATE', '# WINS', '# REGULATORY LOSES', '# SHOOTOUT LOSES']
team_records['# WINS'], team_records['# REGULATORY LOSES'], team_records['# SHOOTOUT LOSES'] = team_records['# WINS'].astype(int), team_records['# REGULATORY LOSES'].astype(int), team_records['# SHOOTOUT LOSES'].astype(int)
team_records
```
